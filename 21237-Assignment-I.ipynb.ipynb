## Something like Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder

for obj_col in data.select_dtypes('object'):
  LE = LabelEncoder()
  data[obj_col] = LE.fit_transform(data[obj_col]) # Converting categorical data to numerical
```

```python
from sklearn.preprocessing import StandardScaler

for num_col in data[['age','bmi']]:
  SS = StandardScaler()
  data[num_col] = SS.fit_transform(data[[num_col]]) # Scaling the numerical data
```

# ****Linear Regression using sklearn****

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[['age','bmi','smoker']]
Y = data['charges']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

print("X_train.shape",X_train.shape)
print("X_test.shape",X_test.shape)
print("Y_train.shape",Y_train.shape)
print("Y_test.shape",Y_test.shape)
```

```python
X_train
```

```python
#load the regression model from sklearn
from sklearn.linear_model import LinearRegression

LR = LinearRegression()

#fit the model based on train samples
LR.fit(X_train, Y_train)


Y_Pred = LR.predict(X_test)

print(Y_Pred)
```

```python
LR.intercept_
```

****# Adjucted R^2****

```python
def adj_r_sqr(X_train,Y_train,X_test,Y_test):
    model = LinearRegression()  
    model.fit(X_train,Y_train)
    r_2_train = model.score(X_train,Y_train)
    r_2_test = model.score(X_test,Y_test)
    adj_r_2_train =  1 - (1-r_2_train)*(len(Y_train)-1)/(len(Y_train)-X_train.shape[1]-1)
    adj_r_2_test =  1 - (1-r_2_test)*(len(Y_test)-1)/(len(Y_test)-X_test.shape[1]-1)  
    return adj_r_2_train,adj_r_2_test
```

```python
adj_r_sqr(X_train,Y_train,X_test,Y_test)
```
